# 2. IMDB'S top rated 100 movies

```python
imdb=urlopen('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
```

```python
sp=BeautifulSoup(imdb,'lxml')
```

```python
movie=sp.find('tbody',class_='lister-list').findAll('tr')
```

```python
movie_list=[]

for mv in movie:
    title=mv.find('td',{'class':'titleColumn'}).find('a').get_text()
    rating=mv.find('td',{'class':'ratingColumn imdbRating'}).find('strong').get_text()
    rating=float(rating)
    year=mv.find('td',{'class':'titleColumn'}).find('span').get_text()
    
    m_list=[title,rating,year]
    movie_list.append(m_list)
```

```python
final_list=movie_list[0:100]
```

```python
# saving in csv file

df=pd.DataFrame(final_list,columns=['movie','rating','year'])
```

```python
df
```

```python
df.to_csv('imdb_top_movies.csv')
```

# 4. Book reviews from Bookpage website

```python
data=urlopen('https://bookpage.com/reviews')
```

```python
sp1=BeautifulSoup(data,'lxml')
```

```python
book=sp1.find_all('div',class_='bp-block article-info')
book
```

```python
book_details=[]

for i in book:
    book_name=i.find('div',class_='flex-article-content').find('a').get_text()
    author_name=i.find('p',class_='sans bold').get_text()
    author_name=author_name[1:-1]
    genre=i.find('p',class_='genre-links hidden-phone').find('a').get_text()
    review=i.find('div',class_='read-full').find('a',i.get('href'))
    
    
    book_list=[book_name,author_name,genre,review]
    book_details.append(book_list[0:5])
    
```

```python
book_details
```

```python
df2=pd.DataFrame(book_details,columns=['Book_name','Author','Genre','Review'])
df2
```

```python
df2.to_csv('book_reviews.csv')
```

```python
soup22
```
