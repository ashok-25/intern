```python
# Lets plot countplots for all features
for i in data.columns:
    train=data.copy()
    sns.countplot(train[i],hue=train['1Gender of respondent'])
    plt.xlabel(i)
    plt.xticks(rotation=90)
    plt.show()
    print('Mode of the {} is : '.format(i) ,train[i].mode())
```
